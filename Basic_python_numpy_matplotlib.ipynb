Dictionary/Map

```python
# Initialization (Literal)
model_config = {
    "learning_rate": 0.001,
    "epochs": 10,
    "optimizer": "Adam"
}

# Adding or Updating
model_config["batch_size"] = 32

# Accessing (Standard)
lr = model_config["learning_rate"]

# Safe Access (Equivalent to Java's getOrDefault)
dropout = model_config.get("dropout_rate", 0.5)

# Iterating through pairs (Most common in ML)
for key, value in model_config.items():
    print(f"Hyperparameter {key} is set to {value}")

# Iterating through keys only
for key in model_config.keys():
    print(key)
```
