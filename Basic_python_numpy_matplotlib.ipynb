Dictionary/Map

```python
# Initialization (Literal)
model_config = {
    "learning_rate": 0.001,
    "epochs": 10,
    "optimizer": "Adam"
}

# Adding or Updating
model_config["batch_size"] = 32

# Accessing (Standard)
lr = model_config["learning_rate"]

# Safe Access (Equivalent to Java's getOrDefault)
dropout = model_config.get("dropout_rate", 0.5)

# Iterating through pairs (Most common in ML)
for key, value in model_config.items():
    print(f"Hyperparameter {key} is set to {value}")

# Iterating through keys only
for key in model_config.keys():
    print(key)
```

```python
import numpy as np
X = np.array([[51,55],[24,19],[0,4]])
print(X[0])
print(X[0][1])
for row in X:
    print(row)
X = X.flatten()
print(X)
print(X[np.array([0,2,4])])
print(X>15)
print(X[X>15])
```

**Matplotlib**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 6, 0.1) #0.1 as unit, generate from 0 to 6
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label = "sin")
plt.plot(x,y2,linestyle = "--", label = "cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
