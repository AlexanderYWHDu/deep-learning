**Automatic Differentiation**

```python
import torch
```

```python
x = torch.arange(4.0)
print(x)
# Can also create x = torch.arange(4.0, requires_grad=True)
x.requires_grad_(True)
print(x.grad) # The gradient is None by default
```

```python
x.grad.zero_()
y = 2 * torch.dot(x, x)
print(y)
y.backward()
print(x.grad)
print(x.grad == 4 * x)
```

```python
x.grad.zero_()
y = x * x
y.backward(gradient=torch.ones(len(y)))  # Faster: y.sum().backward()
x.grad
```

Detaching computation

```python
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.sum().backward()
x.grad
```

```python
x.grad.zero_()
y = x * x
z = y * x

z.sum().backward()
x.grad
```

Gradients and Python Control Flow

```python
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c
```

```python
a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()
a.grad == d / a
```

```python
import matplotlib.pyplot as plt
x = torch.linspace(-2 * torch.pi, 2 * torch.pi, 100, requires_grad=True)

y = torch.sin(x)
y.sum().backward()

x_vals = x.detach().numpy()
y_vals = y.detach().numpy()
dy_vals = x.grad.numpy()

plt.plot(x_vals, y_vals, label='sin(x)')
plt.plot(x_vals, dy_vals, label="sin'(x) via autograd")
plt.legend()
plt.show()
```
