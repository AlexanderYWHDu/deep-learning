**Data manipulation**

```python
import torch
x = torch.arange(12, dtype=torch.float32)
X = x.reshape(3, 4)
X[:2] = 12
x = torch.exp(x)
print(x)
X = torch.arange(12, dtype=torch.float32).reshape((3,4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print(torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1))
print(X == Y)
print(X.sum())
```

Memory

```python
before = id(Y)
Y = Y + X
print(id(Y) == before)
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
before = id(X)
X += Y
id(X) == before
```

torch & numpy

```python
A = X.numpy()
B = torch.from_numpy(A)
print(type(A), type(B))
```

**Automatic Differentiation**

```python
import torch
```

```python
x = torch.arange(4.0)
print(x)
# Can also create x = torch.arange(4.0, requires_grad=True)
x.requires_grad_(True)
print(x.grad) # The gradient is None by default
```

```python
y = 2 * torch.dot(x, x)
print(y)
y.backward()
print(x.grad)
print(x.grad == 4 * x)
```
