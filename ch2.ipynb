**Data manipulation**

```python
import torch
x = torch.arange(12, dtype=torch.float32)
X = x.reshape(3, 4)
X[:2] = 12
x = torch.exp(x)
print(x)
X = torch.arange(12, dtype=torch.float32).reshape((3,4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print(torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1))
print(X == Y)
print(X.sum())
```

Memory

```python
before = id(Y)
Y = Y + X
print(id(Y) == before)
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
before = id(X)
X += Y
id(X) == before
```

torch & numpy

```python
A = X.numpy()
B = torch.from_numpy(A)
print(type(A), type(B))
```
